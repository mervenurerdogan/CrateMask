```python
import cv2 
from google.colab.patches import cv2_imshow
```

```python
path = "/content/drive/MyDrive/MASK/WhatsApp Image 2022-07-27 at 7.36.30 PM.jpeg" # Add the path of the image here
```

```python
img = cv2.imread (path,1)
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
mask1 = cv2.inRange(img, (0, 0, 50), (50, 50,255))


```

```python
cv2_imshow(mask1)
```

```python
cv2.imwrite('mask1_new.jpeg', mask1)
```

```python
## mask of blue color
mask2 = cv2.inRange(img, (50,0,0), (255, 50, 50))
```

```python
cv2_imshow(mask2)
```

```python
cv2.imwrite('mask2_new.jpeg', mask2)
```
